```python
import pandas as pd

# Load the dataset
# Replace 'library_books.csv' with the path to your dataset
library_df = pd.read_csv('/content/mounika revathi .csv')

def display_books():
    """Display all books available in the library."""
    print("Available Books:")
    print(library_df[['book_id', 'title', 'author', 'genre', 'published_year', 'available_copies']])

def search_book(title):
    """Search for a book by title."""
    result = library_df[library_df['title'].str.contains(title, case=False)]
    if result.empty:
        print("Book not found.")
    else:
        print("Search Results:")
        print(result[['book_id', 'title', 'author', 'genre', 'published_year', 'available_copies']])

def borrow_book(book_id):
    """Borrow a book by ID."""
    book = library_df.loc[library_df['book_id'] == book_id]
    if book.empty:
        print("Invalid book ID.")
        return
    copies = book['available_copies'].values[0]
    if copies > 0:
        library_df.loc[library_df['book_id'] == book_id, 'available_copies'] -= 1
        print("Book borrowed successfully.")
    else:
        print("Sorry, this book is currently not available.")

def return_book(book_id):
    """Return a borrowed book by ID."""
    book = library_df.loc[library_df['book_id'] == book_id]
    if book.empty:
        print("Invalid book ID.")
        return
    library_df.loc[library_df['book_id'] == book_id, 'available_copies'] += 1
    print("Book returned successfully.")

# Example usage:
display_books()

# Search for a book
search_book('1984')

# Borrow a book
borrow_book(2)

# Display updated list of books
display_books()

# Return a book
return_book(2)

# Display updated list of books
display_books()
```
